```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d = pd.read_excel("dataset_PCOS.xlsx",sheet_name="Full_new")
d.head().T
```

```python
d.drop('Unnamed: 44',axis=1,inplace=True)
```

```python

d.drop('Patient File No.',axis=1,inplace=True)
d.drop('BMI',axis=1,inplace=True)
d.drop('FSH/LH',axis=1,inplace=True)
d.drop('Waist:Hip Ratio',axis=1,inplace=True)
```

```python
d.drop('Sl. No',axis=1,inplace=True)
```

```python
null_columns = d.columns[d.isnull().any()]
null_columns
```

```python
d['Marraige Status (Yrs)'].fillna(d['Marraige Status (Yrs)'].median(),inplace=True)
d['Fast food (Y/N)'].fillna(d['Fast food (Y/N)'].median(),inplace=True)
```

```python
d.describe().T
```

```python
corr_features=d.corrwith(d["PCOS (Y/N)"]).abs().sort_values(ascending=False)
corr_features=corr_features[corr_features>0.4].index
corr_features
```

```python
d=d[corr_features]
d.head()
```

```python
d.describe().T
```

```python

plt.subplot(1,6,1)
sns.boxplot(x='PCOS (Y/N)',y='Follicle No. (R)',data=d)


plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,6,2)
sns.boxplot(x='PCOS (Y/N)',y='Follicle No. (L)',data=d)

plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,6,3)
sns.boxplot(x='PCOS (Y/N)',y='Skin darkening (Y/N)',data=d)

plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,6,4)
sns.boxplot(x='PCOS (Y/N)',y='hair growth(Y/N)',data=d)

plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,6,5)
sns.boxplot(x='PCOS (Y/N)',y='Weight gain(Y/N)',data=d)

plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,6,6)
sns.boxplot(x='PCOS (Y/N)',y='Cycle(R/I)',data=d)

plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(d.corr(), annot=True)
plt.show()
```
